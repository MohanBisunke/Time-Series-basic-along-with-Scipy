```python
from datetime import datetime
def parse(x):
    return datetime.strptime(x, '%m/%d/%Y')

df = pd.read_csv('https://raw.githubusercontent.com/srivatsan88/YouTubeLI/master/dataset/electricity_consumption.csv', parse_dates = ['Bill_Date'])
df.head()
     
```

```python
df.info()
```

```python
bill_df = df.set_index('Bill_Date')
bill_df
```

```python
df_2018 = bill_df['2016':'2018'][['Billed_amount']]
df_2018
```

Simple moving average = (t + (t-1) + (t-2) + ... + (t-n)) / n

```python
df_2018['sma_rolling3'] = df_2018['Billed_amount'].rolling(window=3).mean()
df_2018
```

```python
df_2018['sma_rolling3_shift1'] = df_2018['Billed_amount'].rolling(window=3).mean().shift(1)
df_2018
```

```python
df_2018.plot()
```

Weighted moving average = (t * weighting factor) + ((t-1) * weighting factor-1) + ((t-n) * weighting factor-n)/n

```python
def wma(weights):
    def calc(x):
        return (weights *x).mean()
    return calc
```

```python
df_2018['WMA3'] = df_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5])))
df_2018
```

```python
df_2018['WMA3S1'] = df_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5]))).shift(1)
df_2018
```

```python
df_2018.plot()
```

```python
df_2018[['Billed_amount' ,'sma_rolling3_shift1','WMA3S1']].plot(figsize=(15,7))
```

Exponential moving average = (Close - previous EMA) * (2/(span+1)) + previous EMA

```python
df_2018['EMA3'] = df_2018['Billed_amount'].ewm(span=3,min_periods=0,adjust=False).mean().shift(1)  #higher span means giving more concerned to old datas
```

```python
df_2018.plot()
```

```python
df_2018[['Billed_amount' ,'sma_rolling3_shift1','WMA3S1','EMA3']].plot(figsize=(15,7))
```

```python
df_2018['EMA5'] = df_2018['Billed_amount'].ewm(span=5,min_periods=0,adjust=False).mean().shift(1)
```

```python
df_2018[['Billed_amount','WMA3S1','EMA3','EMA5']].plot(figsize=(15,7))
```

Exponential Smoothing -
Alpha - 0 to 1 , more value means model is concerned more on present datas.

```python
df_2018['ESMA3'] = df_2018['Billed_amount'].ewm(alpha=0.7,min_periods=3,adjust=False).mean().shift(1)  #alpha and span can't be used once.
df_2018
```

```python
df_2018[['Billed_amount','WMA3S1','EMA3','ESMA3']].plot(figsize=(15,7))
```
